```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import pandas as pd

def load_data(file_path):
    """
    Load data from a CSV file.

    Parameters:
    - file_path (str): Path to the CSV file.

    Returns:
    - pd.DataFrame: Loaded data.
    """
    data = pd.read_csv(file_path)
    return data

def plot_histogram_and_distribution(data, bins=30):
    """
    Create a histogram and fit a normal distribution to the data.

    Parameters:
    - data (pd.DataFrame): Input data.
    - bins (int): Number of bins for the histogram.

    Returns:
    - tuple: Tuple containing (mean, standard deviation) of the fitted normal distribution.
    """
    plt.hist(data, bins=bins, density=True, alpha=0.6, color='b')

    # Fit a normal distribution to the data
    mu, std = norm.fit(data)

    # Plot the PDF
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, mu, std)
    plt.plot(x, p, 'k', linewidth=2)

    return mu, std

def calculate_statistics(data):
    """
    Calculate mean and the value at the 10th percentile of the data.

    Parameters:
    - data (pd.DataFrame): Input data.

    Returns:
    - tuple: Tuple containing (mean, value at the 10th percentile).
    """
    mean_salary = np.mean(data)
    percentile_10 = np.percentile(data, 10)
    return mean_salary, percentile_10

def visualize_results(mean_salary, percentile_10):
    """
    Visualize mean and value at the 10th percentile on the graph.

    Parameters:
    - mean_salary (float): Mean annual salary.
    - percentile_10 (float): Value at the 10th percentile.
    """
    # Plot mean and X on the graph
    plt.axvline(mean_salary, color='r', linestyle='dashed', linewidth=2, label=f'Mean Salary: {mean_salary:.2f}')
    plt.axvline(percentile_10, color='b', linestyle='dashed', linewidth=2, label=f'X (10th percentile): {percentile_10:.2f}')

    # Labeling and title
    plt.xlabel('Annual Salary')
    plt.ylabel('Probability Density')
    plt.title('Probability Density Function of Annual Salaries')
    plt.legend()

    # Show the plot
    plt.show()

def print_results(mean_salary, percentile_10):
    """
    Print calculated values.

    Parameters:
    - mean_salary (float): Mean annual salary.
    - percentile_10 (float): Value at the 10th percentile.
    """
    print(f"Mean Annual Salary (W~): {mean_salary:.2f}")
    print(f"Value X (10th percentile): {percentile_10:.2f}")

def main():
    # Load data from the file located in the same directory
    data = load_data('data5-1.csv')

    # Plot histogram and fit normal distribution
    mu, std = plot_histogram_and_distribution(data)

    # Calculate mean and value at the 10th percentile
    mean_salary, percentile_10 = calculate_statistics(data)

    # Visualize results on the graph
    visualize_results(mean_salary, percentile_10)

    # Print calculated values
    print_results(mean_salary, percentile_10)

if __name__ == "__main__":
    main()
```
